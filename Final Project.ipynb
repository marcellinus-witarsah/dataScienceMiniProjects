Based on the target variable which is not a quantitative variable we might using classification algorithm, there is a lot of classification algorithms which are suited multi-class:

For this one we will use kapasitas_ram, and kapasitas_ram as a feature or predictor variable. While price range is the target variable.

```python
train_data.describe()
```

```python
sns.kdeplot(train_data['daya_baterai'])
sns.kdeplot(train_data['kapasitas_ram'])
```

The standard deviation are relatively high for each columns which might affects our model so we need to scale this. To scale data I will be using 3 class to select which one is the best method for this solution.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer
#prepare data

X = train_data[['daya_baterai', 'px_panjang', 'px_lebar', 'kapasitas_ram']]
X_test = test_data[['daya_baterai', 'px_panjang', 'px_lebar', 'kapasitas_ram']]
y = train_data['price_range']

scalers=[StandardScaler(), MinMaxScaler(), Normalizer()]

scaled_data_train = {}
scaled_data_test = {}

for scaler in scalers:
    scale_data_train = scaler.fit_transform(X)
    scale_data_test = scaler.transform(X_test)
    
    scaled_data_train[scaler.__class__.__name__] = pd.DataFrame(scale_data_train, columns=X.columns)
    scaled_data_test[scaler.__class__.__name__] = pd.DataFrame(scale_data_test, columns=X_test.columns)
```

```python
for scaler in scalers:
    print(f'Train data ({scaler.__class__.__name__})= ')
    display(scaled_data_train[scaler.__class__.__name__])
    print(f'Test data ({scaler.__class__.__name__})= ')
    display(scaled_data_test[scaler.__class__.__name__])  
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
classifiers = [ KNeighborsClassifier(), 
               SVC(probability=True),
               GaussianNB(),
               AdaBoostClassifier(),
               RandomForestClassifier(),
               GradientBoostingClassifier(),
               DecisionTreeClassifier()
              ]

df_algorithm_acc_scaled = pd.DataFrame(columns=['Algorithm', 'Accuracy','Scaler'])
df_algorithm_acc = pd.DataFrame(columns=['Algorithm', 'Accuracy','Scaler'])

for scaler in scalers:
    X_scale = scaled_data_train[scaler.__class__.__name__]
    for c in classifiers:
        cv_score = cross_val_score(c,X_scale,y,cv=10,scoring='f1_micro')
        df_algorithm_acc_scaled= df_algorithm_acc_scaled.append({'Algorithm': c.__class__.__name__, 
                                                   'Accuracy': np.mean(cv_score),
                                                   'Scaler': scaler.__class__.__name__}, 
                                                   ignore_index=True)

for c in classifiers:
    cv_score = cross_val_score(c,X,y,cv=10,scoring='f1_micro')
    df_algorithm_acc= df_algorithm_acc.append({'Algorithm': c.__class__.__name__, 
                                               'Accuracy': np.mean(cv_score),
                                               'Scaler': 'No Scaler'}, 
                                                ignore_index=True)
```

```python
display(df_algorithm_acc)
max_score = df_algorithm_acc.max()[1]
print(df_algorithm_acc[df_algorithm_acc['Accuracy'] == max_score])

display(df_algorithm_acc_scaled)
max_score_scaled = df_algorithm_acc_scaled.max()[1]
print(df_algorithm_acc_scaled[df_algorithm_acc_scaled['Accuracy'] == max_score_scaled])
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Algorithm', y='Accuracy', color='red', data = df_algorithm_acc)
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Algorithm', y='Accuracy', hue='Scaler', data = df_algorithm_acc_scaled)
plt.xticks(rotation=45)
```

Based on the calculation above the algorithm which has the upper hand is the SVC algorithm. After this we will do some hyper parameter tuning for GradientBoostingClassifier() class algorithm using data that has been scaled using StandardScaler() class.

But Without using scaler the accuracy for SVC() is higher.

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV

x_scale_data_train = scaled_data_train['StandardScaler']
gbc = SVC()
parameters = {
    "kernel":['linear','rbf'],
    "gamma": np.arange(0.0001, 100, 20),
    "C": np.arange(0.1, 100, 20)
}

rscv_gbc = RandomizedSearchCV(gbc, param_distributions=parameters, cv=15, scoring='f1_micro')


rscv_gbc.fit(x_scale_data_train, y)
```

```python
rscv_gbc.best_score_
```

Using the best model for predicting smartphone price.

```python
best_model = rscv_gbc.best_estimator_
```

```python
best_model.predict(scaled_data_test['StandardScaler'])
```
